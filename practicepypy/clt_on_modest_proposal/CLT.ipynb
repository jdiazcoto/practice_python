```python
import numpy as np
import pandas as pd 
#Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
list_of_words = []

with open('data.txt','r') as file:
   
    # reading each line    
    for line in file:
   
        # reading each word        
        for word in line.split():
           list_of_words.append(word)
```

```python
#Go ahead and create a list of average letters from the list_of_words
average = 0
list_of_average_letters_in_word = []
for word in list_of_words:
    average = len(word)
    list_of_average_letters_in_word.append(average)
```

```python
def random_samples(population, sample_qty, sample_size):
    sample_means= []
    for i in range(sample_qty):
        sample = population.sample(n=sample_size)
        sample_mean =np.array(sample).mean()
        sample_means.append(sample_mean)
    return sample_means
```

```python
pop = pd.Series(list_of_average_letters_in_word)
samples_from_normal = random_samples(pop, 100,100)
plt.figure(figsize = (7,4))
plt.title("Distribution of Sample Means - Normal Distribution", fontsize=15)
sns.distplot(samples_from_normal, hist=False)
```
